#STOCK PRICE PREDICTOR ON THE BASIS OF THE LAST 10 DAYS RECORD

```python
import numpy as np
import tensorflow as tf
import warnings
warnings.filterwarnings("ignore")

from tensorflow.keras.layers import Dense, SimpleRNN , Embedding
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
```

#TOY_DATA CREATION

```python
time = np.arange(1,100,0.1)
price = np.sin(time)*np.random.randn(len(time))

#normalisation
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
price_norm = scaler.fit_transform(price.reshape(-1,1))
print(price_norm)
```

```python
x,y = [],[]
for i in range(len(price)-1):
  x.append(price[i])
  y.append(price[i+1])


x = np.array(x).reshape(-1,1,1)
y = np.array(y)
```

#MODEL GENERATION

```python
sequence_days = 10
```

```python
model = Sequential([
    SimpleRNN(10,activation = "tanh",input_shape =(sequence_days,1)),
    Dense(1)


])
model.compile(loss = "mse",optimizer=Adam(1e-3),metrics=["mse"])
model.fit(
    x,y,epochs = 20 , validation_split = 0.15
)
```

```python
score = model.evaluate(x,y)
print(f"the loss is : {score[0]},the accurecy is {score[1]}")
```

# Stock Generator

```python
sequence_days = 10
last_sequence = price_norm[-sequence_days:]
last_sequence = last_sequence.reshape(1, sequence_days, 1)
```

```python
next_day_scaled = model.predict(last_sequence)
next_day_price = scaler.inverse_transform(next_day_scaled)

print("Predicted next day's price:", next_day_price[0][0])
```
